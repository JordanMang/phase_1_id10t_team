```python
tnmovie = pd.read_csv('../Data/tn.movie_budgets.csv')
```

```python
tnmovie.head()
```

```python
tnmovie.info()
```

```python
tnmovie['domestic_gross'] = tnmovie['domestic_gross'].str.replace('$','')

tnmovie['production_budget'] = tnmovie['production_budget'].str.replace('$','')

tnmovie['domestic_gross'] = tnmovie['domestic_gross'].str.replace(',','')

tnmovie['production_budget'] = tnmovie['production_budget'].str.replace(',','')
```

```python
tnmovie['worldwide_gross'] = tnmovie['worldwide_gross'].str.replace('$','')

tnmovie['worldwide_gross'] = tnmovie['worldwide_gross'].str.replace(',','')

tnmovie['worldwide_gross'] = tnmovie['worldwide_gross'].astype(int)
```

```python
tnmovie['domestic_gross'] = tnmovie['domestic_gross'].astype(int)

tnmovie['production_budget'] = tnmovie['production_budget'].astype(int)
```

```python
tnmovie.head()
```

```python
tnmovie['profit'] = tnmovie['worldwide_gross'] - tnmovie['production_budget']
```

```python
tnmovie.head()
```

```python
tnmovie = tnmovie.sort_values(by='profit',
                   ascending=False)
```

```python
tnmovie.shape
```

```python
top500 = tnmovie.iloc[0:501]
```

```python
top500
```

```python
moviebasics = pd.read_sql('''
SELECT
    primary_title,
    genres
FROM
    movie_basics
WHERE
    primary_title = original_title

''',conn)
```

```python
moviebasics.head()
```

```python
moviebasics.loc[moviebasics['primary_title'] == 'Titanic']
```

```python

moviebasics.join(top500,
                 on=moviebasics['primary_title'], top500['movie'],
                 how='inner')
```

```python
grouped = pd.merge(moviebasics,
                  tnmovie,
                  left_on='primary_title',
                  right_on='movie')
```

```python
grouped.head()
```

```python
grouped.sort_values(by='genres',
                   ascending=False,
                   inplace=True)
```

```python
grouped.drop_duplicates(subset='primary_title',
                       inplace=True)
```

```python
grouped.sort_values(by='profit',
                   ascending=False,
                   inplace=True)
```
